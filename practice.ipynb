El desconocimiento de fecha no parece estar relacionado con ningÃºn parametro en especial, intentemos solucionarlo de una forma mÃ¡s creativa

```python
def edad_fix(anio):
    if anio==0:
        return None
    return anio
 
artists_billboard['anioNacimiento']=artists_billboard.apply(lambda x: edad_fix(x['anioNacimiento']), axis=1)

def calcula_edad(anio,cuando):
    cad = str(cuando)
    momento = cad[:4]
    if anio==0.0:
        return None
    return int(momento) - anio
 
artists_billboard['edad_en_billboard']=artists_billboard.apply(lambda x: calcula_edad(x['anioNacimiento'],x['chart_date']), axis=1)
```

```python
age_avg = artists_billboard['edad_en_billboard'].mean()
age_std = artists_billboard['edad_en_billboard'].std()
age_null_count = artists_billboard['edad_en_billboard'].isnull().sum()
age_null_random_list = np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count)
 
conValoresNulos = np.isnan(artists_billboard['edad_en_billboard'])
 
artists_billboard.loc[np.isnan(artists_billboard['edad_en_billboard']), 'edad_en_billboard'] = age_null_random_list
artists_billboard['edad_en_billboard'] = artists_billboard['edad_en_billboard'].astype(int)
print("Edad Promedio: " + str(age_avg))
print("DesviÃ³ Std Edad: " + str(age_std))
print("Intervalo para asignar edad aleatoria: " + str(int(age_avg - age_std)) + " a " + str(int(age_avg + age_std)))
```

Tenemos 31 canciones cuya duracion es 0, realicemos lo mismo que con la edad

```python
def duration_fix(duration):
    if duration==0:
        return None
    return duration
 
artists_billboard['durationMin']=artists_billboard.apply(lambda x: duration_fix(x['durationMin']), axis=1)
```

```python
duration_avg = artists_billboard['durationMin'].mean()
duration_std = artists_billboard['durationMin'].std()
print(duration_std)
print(duration_avg)
```

Hay un gran desvÃ­o lo cual deberemos ajustar un poco si queremos evitar alejarnos mucho del valor promedio

```python
duration_std = duration_std * 0.5

duration_null_count = artists_billboard['durationMin'].isnull().sum()
duration_null_random_list = np.random.randint(duration_avg - duration_std, duration_avg + duration_std, size=duration_null_count)
 
conValoresNulos = np.isnan(artists_billboard['durationMin'])
 
artists_billboard.loc[np.isnan(artists_billboard['durationMin']), 'durationMin'] = duration_null_random_list
artists_billboard['durationMin'] = artists_billboard['durationMin'].astype(int)
print("DuraciÃ³n Promedio: " + str(duration_avg))
print("DesvÃ­o Std duraciÃ³n: " + str(duration_std))
print("Intervalo para asignar duraciÃ³n aleatoria: " + str(int(duration_avg - duration_std)) + " a " + str(int(duration_avg + duration_std)))
```

```python
artists_billboard['durationMin'] = artists_billboard['durationMin'].abs()
artists_billboard['durationMin'] = artists_billboard['durationMin'].replace(0, duration_avg)   #Todos los valores deben ser positivos y mayores que 0
```
